```python
import pytesseract
from PIL import Image

pytesseract.pytesseract.tesseract_cmd = r"G:\tools\tesseract.exe"

img_bagus_ocr = Image.open("../dataset/HW_bagus.jpeg")
text_bagus_ocr = pytesseract.image_to_string(img_bagus_ocr)

img_jelek_ocr = Image.open("../dataset/HW_jelek.jpg")
text_jelek_ocr = pytesseract.image_to_string(img_jelek_ocr)

print(text_bagus_ocr)

print(text_jelek_ocr)

```
